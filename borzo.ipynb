# Решение задачи с поддельной монеткой

## Используем метод бинарного поиска

### Шаг 1
- Разделим все монетки на 2 одинаковые кучки: **4 + 4**
- Одну из кучек взвесим на весах (**2 + 2 монетки**)
- Другая останется в руках
**Возможные сценарии:**
- ✅ **Весы отклонились** → поддельная монета на весах
- ⚖️ **Весы уравновешены** → поддельная монета на руках

> Эталонные монеты откладываем в отдельную кучку

---
### Шаг 2
- Кучку из 4 монеток с поддельной разделяем пополам: **2 + 2 монетки**
- На одну чашу весов: **2 монетки из проверяемой кучки**
- На другую чашу весов: **2 эталонные монетки**

**Возможные сценарии:**
- ✅ **Весы отклонились** → поддельная монета на весах
- ⚖️ **Весы уравновешены** → поддельная монета на руках

> Теперь нужно проверить всего **2 монетки**

---
### Шаг 3
- Разделяем кучку из 2 монеток: **1 + 1 монетка**
- На одну чашу весов: **1 монетка из проверяемой кучки**
- На другую чашу весов: **1 эталонная монетка**
**Возможные сценарии:**
- ✅ **Весы отклонились** → поддельная монета на весах
- ⚖️ **Весы уравновешены** → поддельная монета на руках

# Условия и решение задачи доставки заказов

### **Исходные условия:**
- **Расстояние до заказа** имеет равномерное распределение от **0 до 3 км**.
- **200 заказов** поступают с равными интервалами за **480 минут** (с 10 утра до 6 вечера).
- Из этого следует, что:
  - **1 заказ** поступает каждые **2.4 минуты**,
  - В минуту поступает **5/12 заказа**.
- **95% заказов** должны быть выполнены **не более чем за 30 минут**.

---

### **Решение:**

1. **95-й перцентиль расстояния** для всех заказов составляет **2.85 км**.
2. Чтобы **доехать до заказа на этом расстоянии и вернуться** за **1 час**, курьер должен передвигаться со скоростью:
   Скорость = (2.85 * 2) / 1 = 5.7 км/ч
3. За это время поступит **25 заказов**, и первый курьер, который выехал, приедет только к последнему заказу.
4. Если скорость курьера будет выше, количество курьеров можно рассчитать по формуле:
  **(2.85 * 2) / скорость (км/ч) * (60 / 2.4)**
